**Imports and formatting data**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'./datasets/Info_pais.csv', encoding='ISO-8859-1', delimiter=';', decimal=',')
```

**Ejerercicios**

```python
df_order = df.sort_values('Poblacion', ascending=False)
df_order.head(15)
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida']
);
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida']
)
plt.title('Renta per capita Vs. Esperanza de vida')
plt.xlabel('Renta per capita')
plt.ylabel('Esperanza de vida');
```

```python
df_order['Poblacion_normalizada'] = df_order['Poblacion'] / (max(df_order['Poblacion']) / 10000)
df_order
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s = df_order['Poblacion_normalizada']
)
plt.title('Renta per capita Vs. Esperanza de vida')
plt.xlabel('Renta per capita')
plt.ylabel('Esperanza de vida');
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s = df_order['Poblacion_normalizada']
)
plt.title('Renta per capita Vs. Esperanza de vida')
plt.xlabel('Renta per capita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5,10);
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s = df_order['Poblacion_normalizada'],
    c = df_order['Poblacion_normalizada']
)
plt.title('Renta per capita Vs. Esperanza de vida')
plt.xlabel('Renta per capita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5,10);
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s = df_order['Poblacion_normalizada'],
    c = df_order['Poblacion_normalizada']
)
plt.title('Renta per capita Vs. Esperanza de vida')
plt.xlabel('Renta per capita')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5,10)

df_order.reset_index(drop=True, inplace=True)

for i in range(0, 10):
  plt.annotate(
    df_order['País'][i],
      (
        df_order['Renta per capita'][i],
        df_order['Esperanza de vida'][i]
      )
  );
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s = df_order['Poblacion_normalizada'],
    c = df_order['Poblacion_normalizada'],
    alpha = 0.5
)
plt.title('Renta per capita Vs. Esperanza de vida')
plt.xlabel('Renta per capita')
plt.ylabel('Esperanza de vida')
plt.colorbar()

fig = plt.gcf()
fig.set_size_inches(18.5,10)  # para tener el tamanio correcto con el colorbar

df_order.reset_index(drop=True, inplace=True)

for i in range(0, 10):
  plt.annotate(
    df_order['País'][i],
      (
        df_order['Renta per capita'][i],
        df_order['Esperanza de vida'][i]
      )
  );
```

###Conclusion

Después de analizar y retocar los gráficos para su correcta interpretación, podemos ver que existe una correlación entre la renta per capita de un país y su ensperanza de vida. Aunque es innegable esta relación, también podemos observar que países con rentas per capita alta se encuentran con la misma esperanza de vida que países con renta per capita más baja.
Esto se debe a que la generalización que hacemos para esta correlación, no tiene en cuenta otros factores que pueden ser muy determinantes. Ya sea la calidad del agua para la población general o los medicamentos y la sanidad disponibles. Es por ello que creo que es un buen gráfico para hacerse una idea de la correlación, pero no debe de ser el único a tener en cuenta en caso de encontrarte haciendo una investigación.
